## 3. ***Data Wrangling***

### Data Wrangling Code

````python
def properFunction(x):                         # function for converting string into integer
  try:
    num=int(x)
    return int(x)
  except Exception as e:
    return int(x[0])*1000000                       # converting one exception into an proper integer
def SizeFunction(x):
  try:
    return float(x[:-1])                          # converting string into float
  except Exception as e:
    if x=="Varies with device":
      return np.nan                            # replacing 'varies with device' with np.nan
    else:
      try:
         return float(x)                             # converting data into float
      except Exception as e:
          print("Problem here -->",x)
          return x
def InstallFunction(x):
  result=""
  try:
    for i in x:
      if i.isnumeric()==True:
        result=result+i                                      # converting data into integer
    return int(result)
  except:
    return 0
def PriceFunction(x):
  try:
    return int(x)
  except:
     try:
       return float(x[1:])                                      # converting data into float
     except:
       return 0
dataset.drop_duplicates(inplace=True)                              # droping the duplicates
dataset['Reviews']=dataset['Reviews'].apply(properFunction)      # apply the function to convert datatypes
dataset['Size']=dataset['Size'].apply(SizeFunction)             # apply the function to convert datatypes
dataset['Installs']=dataset['Installs'].apply(InstallFunction)  # apply the function to convert datatypes
dataset['Price']=dataset['Price'].apply(PriceFunction)         # apply the function to convert datatypes
dataset=dataset[ dataset['Rating']<=5 ]                        # removing the app which has rating more than 5
````

### What all manipulations have you done and insights you found?

All the manipulations are as follows :-

1.   Convert datatypes of reveiws, size, installs and price from object to integer or float
2.   Droping the duplicates
3.   Removing the apps which are having rating more than 5
